```python
import pandas as pd
import warnings
```

```python
warnings.simplefilter("ignore")
```

```python
path = 'C:\\Users\\Asus VivoBook 11TH\\Downloads\\NOSQL-All_Data.xlsx'
```

```python
df = pd.read_excel(path)
```

```python
unimaj_to_split = df[['Education 1', 'Education FOS 1', 'Education 2', 'Education FOS 2', 'Education 3', 'Education FOS 3']]
```

```python
unimaj_to_split
```

```python
pairs = []
for i in range(1, 4):
    education_col = f'Education {i}'
    education_fos_col = f'Education FOS {i}'
    pairs.append(pd.DataFrame({
        'Education': unimaj_to_split[education_col].dropna(),
        'Education FOS': unimaj_to_split[education_fos_col].dropna()
    }))
```

```python
result_df = pd.concat(pairs, ignore_index=True)
```

```python
result_df
```

```python
result_df = result_df.drop_duplicates().reset_index(drop=True)
```

```python
result_df
```

```python
grouped_df = result_df.groupby('Education')['Education FOS'].agg(lambda x: ', '.join(x)).reset_index()
```

```python
grouped_df = result_df.groupby('Education')['Education FOS'].agg(lambda x: ', '.join(str(val) for val in x if pd.notna(val))).reset_index()
```

```python
cleaned_df = grouped_df.dropna(subset=['Education FOS']).replace('', pd.NA).dropna().reset_index(drop=True)
```

```python
cleaned_df
```

```python
invalid_entries = ['--', '', ' ', 'nan', 'Nan', '2 ANS DE FORMATION PRATIQUE', '6 november', '2004-2005 : Formation dans un', '2013 : Certifié SQL Server 2008 professionnel accordés par l’organisme, (Microsoft']
cleaned_df = cleaned_df[~cleaned_df['Education FOS'].isin(invalid_entries)].reset_index(drop=True)
```

```python
output_path = r'C:\\Users\\Asus VivoBook 11TH\\Desktop\\grouped_data.xlsx'
```

```python
grouped_df.to_excel(output_path, index=False)
```
